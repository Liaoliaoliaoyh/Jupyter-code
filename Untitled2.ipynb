What’s RPN?
 
A Region Proposal Network(PRN) takes an image (of any size) as input and outputs a set of 
rectangular Object proposals , each with an objectness score.
 
What’s the architecture of RPN?
 
This architecture is naturally implemented with an n×n convolutional layer followed
by two sibling 1 × 1 convolutional layers (for reg and cls, respectively).

```python
# 定义anchor的产生函数 generate
def anchor_gen(size_X,size_Y,rpn_stride,scales,ratios):
    # 制作锚点 
    shiftY = np.arange(0, size_Y) * rpn_stride
    shiftX = np.arange(0, size_X) * rpn_stride  # 一乘就映射到原图了
    shiftX,shiftY = np.meshgrid(shiftX,shiftY)  # 组合成网格，锚点
    # 至此，原图的锚点有了

    # 制作锚框
    scales,ratios = np.meshgrid(scales,ratios)
    scales,ratios = scales.flatten(),ratios.flatten()
    scalesY = scales * np.sqrt(ratios)
    scalesX = scales / np.sqrt(ratios)
    # 至此 得到9组锚框的完整尺寸
    # 至此，原图的锚点有了，锚框的尺寸也有了 
    
    # 再组合一遍：在一个锚点上要放9种尺寸的锚框
    centerX,anchorX = np.meshgrid(shiftX,scalesX)
    centerY,anchorY = np.meshgrid(shiftY,scalesY)
    # 至此，把每个框放在了每个锚点上
    
    # 中心点坐标 和 框的尺寸
    anchor_center = np.stack([centerY,centerX],axis=2).reshape(-1,2)
    anchor_size = np.stack([anchorY,anchorX],axis=2).reshape(-1,2)
    
    # 框：返回的是框的左上角坐标和右下角坐标
    boxes = np.concatenate([anchor_center-0.5*anchor_size,anchor_center+0.5*anchor_size],axis=1)
    return boxes
```

```python
anchors = anchor_gen(size_X,size_Y,rpn_stride,scales,ratios)
anchors.shape
```

```python
import numpy as np


# Verify that we compute the same anchors as Shaoqing's matlab implementation:
#
#    >> load output/rpn_cachedir/faster_rcnn_VOC2007_ZF_stage1_rpn/anchors.mat
#    >> anchors
#
#    anchors =
#
#       -83   -39   100    56
#      -175   -87   192   104
#      -359  -183   376   200
#       -55   -55    72    72
#      -119  -119   136   136
#      -247  -247   264   264
#       -35   -79    52    96
#       -79  -167    96   184
#      -167  -343   184   360

# array([[ -83.,  -39.,  100.,   56.],
#       [-175.,  -87.,  192.,  104.],
#       [-359., -183.,  376.,  200.],
#       [ -55.,  -55.,   72.,   72.],
#       [-119., -119.,  136.,  136.],
#       [-247., -247.,  264.,  264.],
#       [ -35.,  -79.,   52.,   96.],
#       [ -79., -167.,   96.,  184.],
#       [-167., -343.,  184.,  360.]])

def generate_anchors(base_size=16, ratios=[0.5, 1, 2],
                     scales=2 ** np.arange(3, 6)):                          # array([3, 4, 5])
    """
    Generate anchor (reference) windows by enumerating aspect ratios X
    scales wrt a reference (0, 0, 15, 15) window.                          # 引用(0,0,15,15)窗口
    """

    base_anchor = np.array([1, 1, base_size, base_size]) - 1
    ratio_anchors = _ratio_enum(base_anchor, ratios)                       # 以基准为16的点对应的基础框 3个
    # array([[[-3.5,  0. ,  2.5],
    #         [ 2. ,  0. , -3. ],
    #         [18.5, 15. , 12.5],
    #         [13. , 15. , 18. ]]])   竖着看 的确是三个框
    anchors = np.vstack([_scale_enum(ratio_anchors[i, :], scales)
                         for i in range(ratio_anchors.shape[0])])
    # array([[-84., -84., -84., -40., -40., -40., 99., 99., 99., 55., 55., 55.],
    #        [-120., -120., -120., -120., -120., -120., 135., 135., 135., 135., 135., 135.],
    #        [-168., -168., -168., -344., -344., -344., 183., 183., 183., 359., 359., 359.]])

    return anchors


def _whctrs(anchor):
    """
    Return width, height, x center, and y center for an anchor (window).
    返回锚点(窗口)的宽度、高度、x中心和y中心。
    """

    w = anchor[2] - anchor[0] + 1             # 16
    h = anchor[3] - anchor[1] + 1             # 16
    x_ctr = anchor[0] + 0.5 * (w - 1)         # 7.5
    y_ctr = anchor[1] + 0.5 * (h - 1)         # 7.5
    return w, h, x_ctr, y_ctr


def _mkanchors(ws, hs, x_ctr, y_ctr):
    """
    Given a vector of widths (ws) and heights (hs) around a center
    (x_ctr, y_ctr), output a set of anchors (windows).
    """

    ws = ws[:, np.newaxis]
    hs = hs[:, np.newaxis]
    anchors = np.hstack((x_ctr - 0.5 * (ws - 1),                    # np.hstack():在水平方向上平铺
                         y_ctr - 0.5 * (hs - 1),
                         x_ctr + 0.5 * (ws - 1),
                         y_ctr + 0.5 * (hs - 1)))
    return anchors
# array([[[-3.5,  0. ,  2.5],
#         [ 2. ,  0. , -3. ],
#         [18.5, 15. , 12.5],
#         [13. , 15. , 18. ]]])   竖着看 的确是三个框啊


def _ratio_enum(anchor, ratios):
    """
    Enumerate a set of anchors for each aspect ratio wrt an anchor.
    为每个长宽比列出一组锚点。
    """

    w, h, x_ctr, y_ctr = _whctrs(anchor)                             # 得到基础anchor的长宽和中心值
    size = w * h                                                     # 256
    size_ratios = size / ratios                                      # [[512. 256. 128.]]
    ws = np.round(np.sqrt(size_ratios))                              # 四舍五入值。 [[23. 16. 11.]]
    hs = np.round(ws * ratios)                                       # 四舍五入值。 [[12. 16. 22.]] 对应[0.5 1 2] 只不过这里有了根号倍
    anchors = _mkanchors(ws, hs, x_ctr, y_ctr)                       # 返回三个以 16×16 的基准框
    return anchors
# array([[[-3.5,  0. ,  2.5],
#         [ 2. ,  0. , -3. ],
#         [18.5, 15. , 12.5],
#         [13. , 15. , 18. ]]])   竖着看 的确是三个框啊

def _scale_enum(anchor, scales):
    """
    Enumerate a set of anchors for each scale wrt an anchor.
    为每个标度列出一组锚点。
    """

    w, h, x_ctr, y_ctr = _whctrs(anchor)
    ws = w * scales
    hs = h * scales
    anchors = _mkanchors(ws, hs, x_ctr, y_ctr)
    return anchors
# array([[ -56.,  -56.,   71.,   71.],
#        [-120., -120.,  135.,  135.],
#        [-248., -248.,  263.,  263.]])

```
